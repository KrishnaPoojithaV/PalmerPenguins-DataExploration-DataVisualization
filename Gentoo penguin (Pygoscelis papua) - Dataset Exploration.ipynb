## Inference from analysis of the dataset

This is the exploratory data analysis of the dataset containing the features of the species, Gentoo penguin (Pygoscelis papua).
General statistics are as follows:

Culmen Length (mm)   : Around 45-50 

Culmen Depth (mm)    : Around 14-16

Flipper Length (mm)  : Around 210-220

Body Mass (g)        : Around 4750-5500

All the species reside in the Biscoe island in the Anvers region.
It is also observed that the male species in general have more Culmen Length, Culmen Depth, FlipperLength and Body Mass than the female species.

There is a slight linear behaviour (inverse) in the values of Delta 15 N (o/oo) and Delta 13 C (o/oo) values, which is the information about the presence of Nitrogen15 and Carbon13 isotopes.
