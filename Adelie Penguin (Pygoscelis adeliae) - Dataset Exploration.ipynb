## Inference from analysis of the dataset

This is the exploratory data analysis of the dataset containing the features of the species, Adelie Penguin (Pygoscelis adeliae). General statistics are as follows:

Culmen Length (mm)   : Around 36-41 


Culmen Depth (mm)    : Around 17-19

Flipper Length (mm)  : Around 185-195

Body Mass (g)        : Around 3300-3900

The species reside in the Dream, Torgersen, Biscoe islands in the Anvers region.
It is also observed that the male species in general have more Culmen Length, Culmen Depth, FlipperLength and Body Mass than the female species.

There is a slight linear behaviour (proportional) in the values of Delta 15 N (o/oo) and Delta 13 C (o/oo) values, which is the information about the presence of Nitrogen15 and Carbon13 isotopes.
